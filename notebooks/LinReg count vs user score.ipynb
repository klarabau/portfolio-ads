# Input
As a starting point for creating models, I created this Linear Regression model on the feature "count". The model and feature were chosen rather randomly to gain first insights into how a model for this dataset works and to get familiar with the data.

```python
# Importing libraries 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from ml import *
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Creating dataframe 
df = pd.read_csv('/datc/nano/notebooks/Target variable & Features (V3).csv')
```

```python
# Specifying x and y
X = df[['Threshold: count']]
y = df[['User score']]
```

```python
# Plotting data
df.plot(kind='scatter', x='Threshold: count', y='User score', title='User score vs. count of particles')
plt.show()
```

```python
# Creating model
model = LinearRegression()
model.fit(X, y)
```

```python
# Visualizing model
df.plot(kind='scatter', x='Threshold: count', y='User score', title='User score vs. count of particles')
plt.scatter(X, model.predict(X), c='orange')
plt.show()
```

```python
# Creating x squared
poly = PolynomialFeatures(degree=2, include_bias=False)
X2 = poly.fit_transform(X)
```

```python
# Creating squared model
model = LinearRegression()
model.fit(X2, y)
```

```python
# Visualizing squared model
df.plot(kind='scatter', x='Threshold: count', y='User score', title='User score vs. count of particles')
plt.scatter(X, model.predict(X2), c='orange')
plt.show()
```
